```python
def problem2(N):
    i = 1
    j = 2
    solution = 2

    while j < N:
        k = i + j
        i = j
        j = k

        if(k % 2 == 0):
            solution += k

    return solution

# Testing
unittest.main(argv=[''], verbosity=2,exit=False)
```
